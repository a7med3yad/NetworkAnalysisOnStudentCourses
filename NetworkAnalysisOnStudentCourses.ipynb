# Library Explanations

- **`pandas as pd`**  
  This library helps you work with tabular data easily. You can read CSV files and manipulate data like in Excel but using Python.

- **`networkx as nx`**  
  Used to create and work with graphs (nodes and edges). Great for visualizing connections or networks.

- **`matplotlib.pyplot as plt`**  
  This is a plotting library. You use it to draw graphs, charts, and other visualizations.

- **`arabic_reshaper`**  
  Arabic text often appears broken in visual outputs. This library reshapes Arabic letters so they connect properly.

- **`get_display from bidi.algorithm`**  
  Ensures Arabic text is displayed from right to left (RTL), which is the correct direction, especially after reshaping.


```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display
```

```python
import pandas as pd
file_path = r"D:\DataScience\modifiedstudentcourse.xlsx"
df = pd.read_excel(file_path,header=1)
```

### Code Explanation:
This function takes an Arabic text string and reshapes it so that it appears correctly when displayed in environments that don't support proper Arabic letter connections (like some Python plotting libraries or UIs).  
- `arabic_reshaper.reshape(text)` reshapes the letters so they connect properly.
- `get_display(reshaped_text)` reverses the text for correct right-to-left display.

This is useful when displaying Arabic text using libraries like `matplotlib` or `pygame`.


```python
def reshape_arabic(text):
    reshaped_text = arabic_reshaper.reshape(text)
    return get_display(reshaped_text)
```

### Code Explanation:
This line groups the DataFrame `df` by the `"Course Code"` column and collects the unique `"Student ID"` values for each course.

- `df.groupby("Course Code")["Student ID"]` groups the data so that for each course code, you get the corresponding student IDs.
- `.apply(set)` converts the list of student IDs for each course into a set (removing duplicates).
- `.to_dict()` turns the grouped data into a dictionary where:
  - **keys** are course codes
  - **values** are sets of student IDs

Useful when you want to quickly access which students are enrolled in each course.

```python
course_students = df.groupby("Course Name (Arabic)")["Student ID"].apply(set).to_dict()
```

### Code Explanation:
This line creates an empty, undirected graph using the `networkx` library and assigns it to the variable `G`.

- `nx.Graph()` creates a **simple undirected graph** (no direction on edges, and no multiple edges or self-loops by default).
- You can now add nodes and edges to `G` using methods like `G.add_node()`, `G.add_edge()`, etc.

This is the starting point for building and analyzing graph data structures.

```python
G = nx.Graph()
```

```python
courses = list(course_students.keys())
for i in range(len(courses)):
    for j in range(i + 1, len(courses)):
        if course_students[courses[i]] & course_students[courses[j]]:
            G.add_edge(courses[i], courses[j])
```

### Code Explanation:
This line of code calculates the number of unique students enrolled in each course.

- `df.groupby("Course Name (Arabic)")`: Groups the DataFrame `df` by the column `"Course Name (Arabic)"`, which contains the names of the courses in Arabic.
- `["Student ID"]`: Focuses on the "Student ID" column to count the unique students.
- `.nunique()`: The `nunique()` function counts the number of unique student IDs in each group (i.e., for each course).
- The result is stored in the `students_per_course` variable, which is a Series containing the number of unique students per course.

This allows you to understand how many distinct students are enrolled in each course.

```python
students_per_course = df.groupby("Course Name (Arabic)")["Student ID"].nunique()
```

### Code Explanation:
This line of code retrieves the top 20 courses with the highest number of unique students.

- `students_per_course.sort_values(ascending=False)`: Sorts the `students_per_course` Series in descending order, so the courses with the most students come first.
- `.head(20)`: Selects the first 20 courses from the sorted list. This gives you the top 20 courses with the highest number of unique students.

The result is stored in the `top_courses` variable, which will contain the names of the top 20 courses along with the number of students enrolled in each.

```python
top_courses = students_per_course.sort_values(ascending=False).head(20)
```

```python
plt.figure(figsize=(14, 6))
top_courses.index = [reshape_arabic(name) for name in top_courses.index]
top_courses.plot(kind="bar", title=reshape_arabic("أكثر 20 مقرر من حيث عدد الطلاب"))

plt.ylabel(reshape_arabic("عدد الطلاب"))
plt.xlabel(reshape_arabic("المقرر"))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Code Explanation:
This line of code calculates the degree of each node (course) in the graph `G`.

- `G.degree()`: This method returns the degree of each node in the graph `G`. The degree of a node refers to the number of edges connected to it (in this case, the number of courses that are connected to a particular course).
- `dict(G.degree())`: Converts the result into a dictionary where the keys are the nodes (courses) and the values are their corresponding degrees (the number of connections to other courses).

The result is stored in the `degrees` variable, which is a dictionary containing the degree of each course in the graph.

```python
degrees = dict(G.degree())
```

### Code Explanation:
This block of code generates a dictionary that maps course codes to their Arabic names and reshapes the course names for better display.

1. **`df.drop_duplicates("Course Code")`:** 
   - Removes any duplicate rows based on the `"Course Code"` column, ensuring that each course code appears only once in the DataFrame.

2. **`[["Course Code", "Course Name (Arabic)"]]`:**
   - Selects only the columns `"Course Code"` and `"Course Name (Arabic)"` from the DataFrame.

3. **`code_to_name_dict = dict(zip(code_to_name["Course Code"], code_to_name["Course Name (Arabic)"]))`:**
   - Creates a dictionary, `code_to_name_dict`, by zipping the `"Course Code"` and `"Course Name (Arabic)"` columns, so each course code is mapped to its corresponding Arabic name.

4. **`reshaped_labels = {code: reshape_arabic(name) for code, name in code_to_name_dict.items()}`:**
   - Iterates through the `code_to_name_dict` and applies the `reshape_arabic` function to reshape the Arabic course names. The reshaped names are then stored in the `reshaped_labels` dictionary, where each key is a course code and each value is the reshaped Arabic name.

This code essentially helps create a dictionary mapping course codes to their reshaped Arabic names.

```python
code_to_name = df.dropna(subset=["Course Code", "Course Name (Arabic)"])
code_to_name = code_to_name.drop_duplicates("Course Code")[["Course Code", "Course Name (Arabic)"]]
code_to_name_dict = dict(zip(code_to_name["Course Code"], code_to_name["Course Name (Arabic)"]))

reshaped_labels = {
    code: reshape_arabic(name)
    for code, name in code_to_name_dict.items()
}
```

```python
sorted_degrees = dict(sorted(degrees.items(), key=lambda x: x[1], reverse=True))
top_degrees = dict(list(sorted_degrees.items())[:20])
labels = [reshaped_labels.get(code, code) for code in top_degrees.keys()]

plt.figure(figsize=(14, 6))
plt.bar(labels, top_degrees.values(), color='mediumseagreen')

plt.title(reshape_arabic("أعلى 20 مقرر من حيث درجة التداخل (عدد المواد المشتركة بالطلاب)"), fontsize=14)
plt.ylabel(reshape_arabic("الدرجة"), fontsize=12)
plt.xlabel(reshape_arabic("المقرر"), fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Code Explanation:
This block of code uses a greedy coloring algorithm to assign "colors" (representing periods or timeslots) to the nodes (courses) in the graph `G`, ensuring that no two connected nodes (courses) share the same color.

1. **`nx.coloring.greedy_color(G, strategy='largest_first')`:**
   - This function applies a greedy coloring algorithm to the graph `G`.
   - The `strategy='largest_first'` argument specifies that the algorithm should color the nodes starting with the largest (most connected) nodes first.
   - It returns a dictionary where the keys are the nodes (courses) and the values are the "colors" (periods) assigned to those nodes.

2. **`num_periods = max(colors.values()) + 1`:**
   - Finds the maximum color value (which represents the last period used) and adds 1 to determine the total number of periods (timeslots) required.

3. **`print(f"عدد الفترات المطلوبة: {num_periods}")`:**
   - Prints the number of periods required, based on the number of unique colors assigned to the nodes.

The result gives the number of distinct periods (or timeslots) needed to schedule all the courses while ensuring no two connected courses are scheduled at the same time.

```python
colors = nx.coloring.greedy_color(G, strategy='largest_first')
num_periods = max(colors.values()) + 1
print(f"عدد الفترات المطلوبة: {num_periods}")
```

### Code Explanation:
This block of code creates an exam schedule DataFrame that associates each course code with its assigned period (or timeslot), based on the coloring results from the previous step.

1. **`exam_schedule = pd.DataFrame.from_dict(colors, orient='index', columns=['Period'])`:**
   - Converts the `colors` dictionary (which maps course codes to their assigned periods) into a Pandas DataFrame.
   - `orient='index'` means that the keys of the dictionary (`course codes`) will be used as the index of the DataFrame.
   - `columns=['Period']` specifies that the column name for the periods is `"Period"`.

2. **`exam_schedule.index.name = 'Course Code'`:**
   - Renames the index of the DataFrame to `"Course Code"`, making it clear that the index represents the course codes.

3. **`exam_schedule = exam_schedule.reset_index()`:**
   - Resets the index of the DataFrame and converts it into a regular column, so the course codes are now part of the DataFrame as a normal column (instead of the index).

This code generates a table that shows each course code and its corresponding exam period.

```python
# lets start coloring the graph
import time

strategies = [
    "largest_first", # color largest deg nodes first
    "random_sequential", # just color nodes in random order without considering deg or anything else
    "independent_set", # taking nodes in independent set then remove the largest set first, ensures no conflict, heigher complexity
    "connected_sequential_bfs", # basic bfs coloring
    "connected_sequential_dfs", # basic dfs coloring
    "DSATUR"
]

results = {}

for strategy in strategies:
    
    start = time.time()
    coloring = nx.coloring.greedy_color(G, strategy=strategy)
    end = time.time()

    # use set to avoid duplicated colors
    n_colors = len(set(coloring.values()))
    
    # exam period ===> min number of colors needed by the alogorithm
    results[strategy] = {
        "periods": n_colors,
        "time_needed": round(end - start, 4),
        "coloring": coloring
    }

print("🎨 Comparison of Coloring Strategies:")
print("Strategy\t\tPeriods\t\tTime (s)")
print("----------------------------------------")

for strategy, data in results.items():
    print(f"{strategy:<25}{data['periods']:<12}{data['time_needed']}")
```
